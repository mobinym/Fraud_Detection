````python
# TODO: ุดูุงุณุง ุฏุงุฏู ูุง ูพุฑุช ุจู ฺฉูฺฉ IQR
# ุชุงุจุน ุจุฑุง ุดูุงุณุง ุฏุงุฏูโูุง ูพุฑุช ุจุง ุงุณุชูุงุฏู ุงุฒ IQR
def detect_outliers_iqr(df):
    outlier_counts = {}
    for column in df.select_dtypes(include=np.number).columns:  # ููุท ูฺฺฏโูุง ุนุฏุฏ
        Q1 = df[column].quantile(0.25)  # ฺุงุฑฺฉ ุงูู
        Q3 = df[column].quantile(0.75)  # ฺุงุฑฺฉ ุณูู
        IQR = Q3 - Q1  # ูุญุงุณุจู IQR
        lower_bound = Q1 - 1.5 * IQR  # ุขุณุชุงูู ูพุงู
        upper_bound = Q3 + 1.5 * IQR  # ุขุณุชุงูู ุจุงูุง
        outliers = (df[column] < lower_bound) | (df[column] > upper_bound)  # ุชุนู ุฏุงุฏูโูุง ูพุฑุช
        outlier_counts[column] = outliers.sum()  # ุดูุงุฑุด ุชุนุฏุงุฏ ุฏุงุฏูโูุง ูพุฑุช

    return outlier_counts

# ูุฑุงุฎูุงู ุชุงุจุน ู ููุงุด ุชุนุฏุงุฏ ุฏุงุฏูโูุง ูพุฑุช
outliers_iqr = detect_outliers_iqr(df_copy)
print("ุชุนุฏุงุฏ ุฏุงุฏูโูุง ูพุฑุช ุจุง ุงุณุชูุงุฏู ุงุฒ IQR:")
print(outliers_iqr)
````

````python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

def impute_missing_data(df, strategy_dict=None, auto_impute=True):
    """
    ุฏุงุฏูโูุง ฺฏูุดุฏู ุฑู ุจุง ุงุณุชุฑุงุชฺโูุง ูุดุฎุต ุง ุจู ุตูุฑุช ุฎูุฏฺฉุงุฑ ูพุฑ ูโฺฉูู.
    ูพุงุฑุงูุชุฑูุง:
        df: ุฏุชุงูุฑู ูุฑูุฏ
        strategy_dict: ุฏฺฉุดูุฑ ุงุฒ ุณุชููโูุง ู ุงุณุชุฑุงุชฺโูุง (ูุซูุงู {'age': 'mean', 'job': 'mode'})
        auto_impute: ุงฺฏู True ุจุงุดูุ ุจู ุตูุฑุช ุฎูุฏฺฉุงุฑ ุจุฑ ุงุณุงุณ ููุน ุฏุงุฏู ูพุฑ ูโฺฉูู
    """
    
    # ูพุฑ ฺฉุฑุฏู ุฎูุฏฺฉุงุฑ
    if auto_impute:
        for column in df.columns:
            if df[column].dtype == 'object':  # ุณุชููโูุง ุฑุดุชูโุง
                mode_value = df[column].mode()[0]
                df[column] = df[column].fillna(mode_value)  # ูพุฑ ฺฉุฑุฏู ุจุง ูุฏ
            else:  # ุณุชููโูุง ุนุฏุฏ
                mean_value = df[column].mean()
                df[column] = df[column].fillna(mean_value)  # ูพุฑ ฺฉุฑุฏู ุจุง ูุงูฺฏู
    
    # ูพุฑ ฺฉุฑุฏู ุจุง ุงุณุชุฑุงุชฺ ูุดุฎุต
    if strategy_dict is not None:
        for column, strategy in strategy_dict.items():
            if strategy == 'mean':
                mean_value = df[column].mean()
                df[column] = df[column].fillna(mean_value)
            elif strategy == 'median':
                median_value = df[column].median()
                df[column] = df[column].fillna(median_value)
            elif strategy == 'mode':
                mode_value = df[column].mode()[0]
                df[column] = df[column].fillna(mode_value)
            elif strategy == 'knn':
                # ุงุณุชูุงุฏู ุงุฒ KNNImputer ุจุฑุง ูพุฑ ฺฉุฑุฏู ููุงุฏุฑ ฺฏูุดุฏู
                knn_imputer = KNNImputer(n_neighbors=5)
                df[column] = knn_imputer.fit_transform(df[[column]])
            elif strategy == 'unknown':
                df[column] = df[column].fillna('unknown')

    return df

# ุชุนุฑู ุงุณุชุฑุงุชฺโูุง ุจุฑ ุงุณุงุณ ููุงุฏุฑ ฺฏูุดุฏู ุฏุฑ ุฏุชุงูุฑู ุดูุง
custom_strategies = {
    'Transaction_Type': 'mode',           # ูพุฑ ฺฉุฑุฏู ุจุง ูุฏ
    'Device_Type': 'mode',                # ูพุฑ ฺฉุฑุฏู ุจุง ูุฏ
    'Card_Type': 'mode',                  # ูพุฑ ฺฉุฑุฏู ุจุง ูุฏ
    'Authentication_Method': 'mode',      # ูพุฑ ฺฉุฑุฏู ุจุง ูุฏ
    'Timestamp': 'mode',                  # ูพุฑ ฺฉุฑุฏู ุจุง ูุฏ (ุงุฎุชุงุฑ)
    'Year': 'mode',                       # ูพุฑ ฺฉุฑุฏู ุจุง ูุฏ (ุงุฎุชุงุฑ)
    'Month': 'mode',                      # ูพุฑ ฺฉุฑุฏู ุจุง ูุฏ (ุงุฎุชุงุฑ)
    'Day': 'mode',                        # ูพุฑ ฺฉุฑุฏู ุจุง ูุฏ (ุงุฎุชุงุฑ)
    'Day_of_Week': 'mode',                # ูพุฑ ฺฉุฑุฏู ุจุง ูุฏ (ุงุฎุชุงุฑ)
    'Hour': 'mode'                        # ูพุฑ ฺฉุฑุฏู ุจุง ูุฏ (ุงุฎุชุงุฑ)
}

# ุงุนูุงู ุชุงุจุน
df_processed = impute_missing_data(df_cleaned, strategy_dict=custom_strategies, auto_impute=False)

# ุจุฑุฑุณ ูุชุงุฌ

df_processed.isnull().sum()
````
