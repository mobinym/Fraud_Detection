````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# فرض بر این است که دیتافریم شما به نام df است
# بررسی توزیع کلاس‌ها در متغیر هدف
class_distribution = df['Fraud_Label'].value_counts()

# نمایش توزیع کلاس‌ها
print(class_distribution)

# رسم نمودار توزیع کلاس‌ها
plt.figure(figsize=(8, 5))
sns.barplot(x=class_distribution.index, y=class_distribution.values, palette='viridis')
plt.title('Class Distribution in Target Variable (Fraud_Label)')
plt.xlabel('Fraud_Label')
plt.ylabel('Number of Samples')
plt.xticks(ticks=[0, 1], labels=['Non-Fraudulent (0)', 'Fraudulent (1)'])
plt.show()
````

````python
# TODO: متد زیر را برای رسم توزیع داده ها کامل کنید.
import matplotlib.pyplot as plt
import seaborn as sns

def plot_numeric_distributions(df):
    # تعیین اندازه شکل
    plt.figure(figsize=(15, 10))
    
    # لیست ویژگی‌های عددی در دیتافریم
    numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
    
    # حلقه برای رسم توزیع هر ویژگی عددی
    for i, column in enumerate(numeric_columns, 1):
        plt.subplot(4, 4, i)  # تنظیم تعداد زیرنمودارها
        sns.histplot(df[column], kde=True, bins=30, color='blue')  # رسم هیستوگرام با نمودار KDE
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
    
    plt.tight_layout()  # تنظیم فاصله بین زیرنمودارها
    plt.show()  # نمایش نمودارها
````

````python
# TODO: شناسایی داده های پرت به کمک IQR
# تابع برای شناسایی داده‌های پرت با استفاده از IQR
def detect_outliers_iqr(df):
    outlier_counts = {}
    for column in df.select_dtypes(include=np.number).columns:  # فقط ویژگی‌های عددی
        Q1 = df[column].quantile(0.25)  # چارک اول
        Q3 = df[column].quantile(0.75)  # چارک سوم
        IQR = Q3 - Q1  # محاسبه IQR
        lower_bound = Q1 - 1.5 * IQR  # آستانه پایین
        upper_bound = Q3 + 1.5 * IQR  # آستانه بالا
        outliers = (df[column] < lower_bound) | (df[column] > upper_bound)  # تعیین داده‌های پرت
        outlier_counts[column] = outliers.sum()  # شمارش تعداد داده‌های پرت

    return outlier_counts

# فراخوانی تابع و نمایش تعداد داده‌های پرت
outliers_iqr = detect_outliers_iqr(df_copy)
print("تعداد داده‌های پرت با استفاده از IQR:")
print(outliers_iqr)
````

````python
# TODO: شناسایی داده های پرت به کمک Z-Score
import numpy as np
import pandas as pd

# تابع برای شناسایی داده‌های پرت با استفاده از Z-Score
def detect_outliers_z_score(df):
    outlier_counts = {}
    for column in df.select_dtypes(include=np.number).columns:  # فقط ویژگی‌های عددی
        z_scores = (df[column] - df[column].mean()) / df[column].std()  # محاسبه Z-Score
        outliers = np.abs(z_scores) > 3  # تعیین آستانه 3
        outlier_counts[column] = outliers.sum()  # شمارش تعداد داده‌های پرت

    return outlier_counts

# فراخوانی تابع و نمایش تعداد داده‌های پرت
outliers_z_score = detect_outliers_z_score(df_copy)
print("تعداد داده‌های پرت با استفاده از Z-Score:")
print(outliers_z_score)
````

````python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

def impute_missing_data(df, strategy_dict=None, auto_impute=True):
    """
    داده‌های گمشده رو با استراتژی‌های مشخص یا به صورت خودکار پر می‌کنه.
    پارامترها:
        df: دیتافریم ورودی
        strategy_dict: دیکشنری از ستون‌ها و استراتژی‌ها (مثلاً {'age': 'mean', 'job': 'mode'})
        auto_impute: اگه True باشه، به صورت خودکار بر اساس نوع داده پر می‌کنه
    """
    
    # پر کردن خودکار
    if auto_impute:
        for column in df.columns:
            if df[column].dtype == 'object':  # ستون‌های رشته‌ای
                mode_value = df[column].mode()[0]
                df[column] = df[column].fillna(mode_value)  # پر کردن با مد
            else:  # ستون‌های عددی
                mean_value = df[column].mean()
                df[column] = df[column].fillna(mean_value)  # پر کردن با میانگین
    
    # پر کردن با استراتژی مشخص
    if strategy_dict is not None:
        for column, strategy in strategy_dict.items():
            if strategy == 'mean':
                mean_value = df[column].mean()
                df[column] = df[column].fillna(mean_value)
            elif strategy == 'median':
                median_value = df[column].median()
                df[column] = df[column].fillna(median_value)
            elif strategy == 'mode':
                mode_value = df[column].mode()[0]
                df[column] = df[column].fillna(mode_value)
            elif strategy == 'knn':
                # استفاده از KNNImputer برای پر کردن مقادیر گمشده
                knn_imputer = KNNImputer(n_neighbors=5)
                df[column] = knn_imputer.fit_transform(df[[column]])
            elif strategy == 'unknown':
                df[column] = df[column].fillna('unknown')

    return df

# تعریف استراتژی‌ها بر اساس مقادیر گمشده در دیتافریم شما
custom_strategies = {
    'Transaction_Type': 'mode',           # پر کردن با مد
    'Device_Type': 'mode',                # پر کردن با مد
    'Card_Type': 'mode',                  # پر کردن با مد
    'Authentication_Method': 'mode',      # پر کردن با مد
    'Timestamp': 'mode',                  # پر کردن با مد (اختیاری)
    'Year': 'mode',                       # پر کردن با مد (اختیاری)
    'Month': 'mode',                      # پر کردن با مد (اختیاری)
    'Day': 'mode',                        # پر کردن با مد (اختیاری)
    'Day_of_Week': 'mode',                # پر کردن با مد (اختیاری)
    'Hour': 'mode'                        # پر کردن با مد (اختیاری)
}

# اعمال تابع
df_processed = impute_missing_data(df_cleaned, strategy_dict=custom_strategies, auto_impute=False)

# بررسی نتایج

df_processed.isnull().sum()
````

````python
# TODO: متوازن سازی کلاس هدف

from imblearn.over_sampling import SMOTE, ADASYN
from imblearn.combine import SMOTETomek, SMOTEENN
from imblearn.under_sampling import RandomUnderSampler
from sklearn.metrics import classification_report, confusion_matrix

# روش‌های متوازن‌سازی
sampling_methods = {
    'SMOTE': SMOTE(random_state=42),
    'SMOTETomek': SMOTETomek(random_state=42),
    'SMOTEENN': SMOTEENN(random_state=42),
    'ADASYN': ADASYN(random_state=42),
    'RandomUnderSampler': RandomUnderSampler(random_state=42)
}

# ارزیابی هر روش
for method_name, method in sampling_methods.items():
    print(f"---{method_name}---")
    
    # اعمال روش متوازن‌سازی فقط روی داده‌های آموزشی
    X_train_resampled, y_train_resampled = method.fit_resample(X_train_scaled, y_train)
    
    # آموزش مدل با داده‌های متوازن‌شده (به عنوان مثال Gradient Boosting)
    model = GradientBoostingClassifier(random_state=42)
    model.fit(X_train_resampled, y_train_resampled)
    
    # پیش‌بینی روی داده‌های آزمایشی
    y_pred = model.predict(X_test_scaled)
    
    # ارزیابی عملکرد
    print(classification_report(y_test, y_pred))
    print(confusion_matrix(y_test, y_pred))
    print("\n")
````

````python
# TODO: رسم نمودار ROC

from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# محاسبه احتمالات پیش‌بینی (Probabilities) برای داده‌های آزمایشی
y_test_prob = best_model.predict_proba(X_test_scaled)[:, 1]

# محاسبه مقادیر FPR (False Positive Rate)، TPR (True Positive Rate) و Thresholds
fpr, tpr, thresholds = roc_curve(y_test, y_test_prob)

# محاسبه AUC
auc_score = roc_auc_score(y_test, y_test_prob)

# رسم نمودار ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {auc_score:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--', label='Random Classifier')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.grid()
plt.show()
````

### **تحلیل نمودار ROC و ارزیابی مدل**

---

### **1. مقدار AUC**
- **AUC = 0.99**: این مقدار نشان‌دهنده عملکرد بسیار عالی مدل است. مقدار AUC نزدیک به 1 نشان می‌دهد که مدل توانایی بسیار خوبی در تفکیک کلاس‌ها دارد.

---

### **2. شکل نمودار**
- **نزدیکی به گوشه بالا سمت چپ**: منحنی ROC در نزدیکی گوشه بالا سمت چپ قرار دارد. این نشان‌دهنده نرخ مثبت درست (**True Positive Rate**) بالا و نرخ مثبت کاذب (**False Positive Rate**) پایین است، که بسیار مطلوب است.
- **خط قرمز (Random Classifier)**: این خط نشان‌دهنده عملکرد یک مدل تصادفی است (AUC = 0.5). منحنی مدل شما بسیار بالاتر از این خط قرار دارد، که نشان‌دهنده برتری واضح مدل نسبت به یک مدل تصادفی است.

---

### **3. تفسیر نرخ‌ها**
- **False Positive Rate (FPR)**: مدل توانسته نرخ مثبت کاذب را در بیشتر نقاط نمودار پایین نگه دارد. این یعنی مدل کمتر تراکنش‌های غیرکلاهبرداری را به اشتباه به عنوان کلاهبرداری شناسایی کرده است.
- **True Positive Rate (TPR)**: مدل نرخ مثبت درست بالایی دارد، به این معنی که اکثر تراکنش‌های کلاهبرداری را به درستی شناسایی کرده است.

---

### **4. عملکرد مدل**
- **حساسیت (Sensitivity/Recall)**: مدل توانسته تراکنش‌های کلاهبرداری را با نرخ بسیار بالایی شناسایی کند، که برای این مسئله بسیار مهم است.
- **اختصاصی بودن (Specificity)**: نرخ مثبت کاذب پایین نشان‌دهنده توانایی مدل در جلوگیری از شناسایی اشتباه تراکنش‌های غیرکلاهبرداری است.
